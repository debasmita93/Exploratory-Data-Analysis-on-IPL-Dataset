# In Different IPL Seasons the number of players are different. Thus merging the dataframes in such a way that records will be combined

```python
numeric_columns = [
    'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score', 'Batting_Average',
    'Balls_Faced', 'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours',
    'Sixes', 'Catches_Taken', 'Stumpings', 'Matches_Bowled', 'Balls_Bowled',
    'Runs_Conceded', 'Wickets_Taken', 'Bowling_Average', 'Economy_Rate',
    'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls','Year'
]
```

```python
for col in numeric_columns:
    Data_2024[col] = pd.to_numeric(Data_2024[col], errors='coerce')
    Data_2023[col] = pd.to_numeric(Data_2023[col], errors='coerce')
    Data_2022[col] = pd.to_numeric(Data_2022[col], errors='coerce')
    Data_2021[col] = pd.to_numeric(Data_2021[col], errors='coerce')
    Data_2020[col] = pd.to_numeric(Data_2020[col], errors='coerce')
```

```python
dataframes = [Data_2024, Data_2023, Data_2022, Data_2021, Data_2020]
```

```python
columns_to_sum = [
    'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Balls_Faced', 'Centuries',
    'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings',
    'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken',
    'Four_Wicket_Hauls', 'Five_Wicket_Hauls'
]
columns_to_max = ['Highest_Score']
```

```python
combined_df = dataframes[0]
```

```python
combined_df
```

```python

for df in dataframes:
    for col in columns_to_sum + columns_to_max:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Merge DataFrames and compute required statistics
for df in dataframes[1:]:
    combined_df = pd.merge(combined_df, df, on='Player_Name', how='outer', suffixes=('', '_y'))

    # Sum specified columns across all years
    for col in columns_to_sum:
        combined_df[col] = combined_df[col].fillna(0) + combined_df[f'{col}_y'].fillna(0)

    # Find the maximum for Highest_Score
    for col in columns_to_max:
        combined_df[col] = combined_df[[col, f'{col}_y']].max(axis=1)

    # Update the Year column to the latest year
    combined_df['Year'] = combined_df[['Year', f'Year_y']].max(axis=1)

    # Drop unnecessary columns after merging
    combined_df.drop(columns=[f'{col}_y' for col in columns_to_sum + columns_to_max] + [f'Year_y'], inplace=True)

# Calculate Batting_Average as total Runs_Scored / total Matches_Batted
combined_df['Batting_Average'] = combined_df['Runs_Scored'] / combined_df['Matches_Batted']

# Calculate Bowling_Average as total Runs_Conceded / total Wickets_Taken
combined_df['Bowling_Average'] = combined_df['Runs_Conceded'] / combined_df['Wickets_Taken']

# Calculate Economy_Rate as total Runs_Conceded / (Balls_Bowled / 6)
combined_df['Economy_Rate'] = combined_df['Runs_Conceded'] / (combined_df['Balls_Bowled'] / 6)

# Calculate Batting_Strike_Rate as (Runs_Scored / Balls_Faced) * 100
combined_df['Batting_Strike_Rate'] = (combined_df['Runs_Scored'] / combined_df['Balls_Faced']) * 100

combined_df['Bowling_Strike_Rate'] = combined_df['Balls_Bowled'] / combined_df['Wickets_Taken']

combined_df['Batting_Average'].fillna(0, inplace=True)
combined_df['Bowling_Average'].fillna(0, inplace=True)
combined_df['Economy_Rate'].fillna(0, inplace=True)
combined_df['Batting_Strike_Rate'].fillna(0, inplace=True)
combined_df['Bowling_Strike_Rate'].fillna(0, inplace=True)

# Original Columns to Keep
final_columns = [
    'Player_Name', 'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score',
    'Batting_Average', 'Balls_Faced', 'Batting_Strike_Rate', 'Centuries',
    'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings',
    'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken',
    'Best_Bowling_Match', 'Bowling_Average', 'Economy_Rate',
    'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls'
]

combined_df = combined_df[final_columns]
```

```python
def plot_stats(data , x_col , y_col , title , xlabel , ylabel , palette):
  plt.figure(figsize=(10 , 6))
  sns.barplot(data = data.sort_values(by = y_col , ascending=False),
              x = x_col , y = y_col , palette = palette)
  plt.xticks(rotation = 45 , ha = 'right')
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.title(title)

  plt.tight_layout()
  plt.show()
```
